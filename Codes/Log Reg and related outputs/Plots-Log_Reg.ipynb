```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.gridspec as gridspec
import seaborn as sns
```

# RQ1

## AUC

```python
df = pd.read_csv('output_Log_Reg/auc_all_mat_6_jdt_LR.csv',index_col=0)
round(df,2)
df2 = pd.read_csv('output_Log_Reg/auc_last_mat_6_jdt_LR.csv',index_col=0)
round(df2,2)
plt.rcParams.update({'font.size': 12})
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2,figsize=(20,8))
plt.subplots_adjust(wspace=0, hspace=0)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
cmap = sns.diverging_palette(220, 10, as_cmap=True)

with sns.axes_style("white"):
    ax = sns.heatmap(round(df,2), #vmax=.3, 
                     square=True,
                     annot=True,
                     annot_kws={"size": 10,"weight": "bold"},
                     linewidths=.5,
                     ax=ax2,
                     yticklabels=False,
                     cmap=cmap)
    ax.set_xticklabels(range(1, df.shape[1]+1), fontsize=15)    
    ax.set_aspect('equal')
    ax.text(0.05, 0.95, "Long Period", transform=ax.transAxes, fontsize=25,
        verticalalignment='top', bbox=props)
    
with sns.axes_style("white"):
    ax = sns.heatmap(round(df2,2), #vmax=.3, 
                     square=True,
                     annot=True,
                     annot_kws={"size": 10,"weight": "bold"},
                     linewidths=.5, 
                     ax=ax1,
                     cmap=cmap,
                     cbar=False)
    ax.set_ylabel("Six-Month Training Period Breakdown",fontsize=18)
    ax.set_aspect('equal')
    ax.set_yticklabels(list(reversed(range(1, df.shape[1]+1))), fontsize=15)
    ax.set_xticklabels(range(1, df.shape[1]+1), fontsize=15)    
    plt.text(1.2, -0.12, "Six-Month Testing Period Breakdown",
         horizontalalignment='center',
         fontsize=18,
         transform = ax1.transAxes)
    ax.text(0.05, 0.95, "Short Period", transform=ax.transAxes, fontsize=25,
        verticalalignment='top', bbox=props)
```
